## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Varibles

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
```

## Add priceUSD column

```python
df = df.assign(priceUSD = lambda priceUSD: df['price'] / 4)
df
```

## Create graph

```python

df["priceUSD"].plot(title = "Price of goods (USD)", color = "red", linestyle = 'dashed', grid = True)
```
