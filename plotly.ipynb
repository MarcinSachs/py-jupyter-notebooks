# Showing salaries graph using plotly

## Imports

```python
import plotly.graph_objects as go
```

## Varibles

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]


layout = {
	'title': 'Salaries'
}

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
```

## Graph

```python

fig = go.Figure(go.Bar(
            x=names,
            y=salary_values,
            orientation='v'),layout)

fig.show()
```
